```python
%matplotlib inline

from matplotlib import pyplot as plt
from matplotlib import patches as mpatches
import numpy as np
import pandas as pd
import math
import sys

import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
def solve_system(data):
    n = len(data)
    for i in range(0, n - 1):
        assert (abs(data[i][0]) < 1e-5)
        cof = data[i + 1][0] / data[i][1]
        data[i + 1][0] -= data[i][1] * cof
        data[i + 1][1] -= data[i][2] * cof
        data[i + 1][3] -= data[i][3] * cof

    answer = [0] * n

    answer[-1] = data[-1][3] / data[-1][1]
    for i in range(n - 2, -1, -1):
        answer[i] = (data[i][3] - answer[i + 1] * data[i][2]) / data[i][1]
        
    return answer
```

```python
def initHill(maxX, dx):
    res = []
    curX = -maxX
    while curX <= maxX:
        res.append(math.exp(- (curX * curX)))
        curX += dx
    return res
```

```python
def leapfrog(maxX, dx, maxT, dt, u, k, initAlgo):
    s = u * dt / dx
    r = k * dt / (dx * dx)
    init = initHill(maxX, dx)
    answer = []
    answer.append(init)
    answer.append(init)
    curT = dt * 2
    l = len(init)
    while (curT <= maxT):
        curRes = [0] * len(init)
        curRes[0] = answer[-1][0]
        curRes[-1] = answer[-1][-1]
        for i in range(1, l - 1):
            A = answer[-2][i]
            B = -s * (answer[-1][i + 1] - answer[-1][i - 1])
#             C = 2 * r * (answer[-1][i - 1] - 2 * answer[-1][i] + answer[-1][i + 1])
            curRes[i] = A + B #+ C

        answer.append(curRes)
        curT += dt
    return answer
```

```python
def explicit_upstream(maxX, dx, maxT, dt, u, k, initAlgo):
    s = u * dt / dx
    r = k * dt / (dx * dx)
    
    init = initAlgo(maxX, dx)
    answer = []
    answer.append(init)
    answer.append(init)
    
    curT = dt * 2
    l = len(init)
    while (curT <= maxT):
        curRes = [0] * len(init)
        curRes[0] = answer[-1][0]
        curRes[-1] = answer[-1][-1]
        for i in range(1, l - 1):
            A = answer[-1][i]
            B = -s * (answer[-1][i + 1] - answer[-1][i])
            C = r * (answer[-1][i - 1] - 2 * answer[-1][i] + answer[-1][i + 1])
            curRes[i] = A + B + C

        answer.append(curRes)
        curT += dt
    return answer
```

```python
def explicit_downstream(maxX, dx, maxT, dt, u, k, initAlgo):
    s = u * dt / dx
    r = k * dt / (dx * dx)
    init = initAlgo(maxX, dx)
    answer = []
    answer.append(init)
    answer.append(init)
    curT = dt * 2
    l = len(init)
    while (curT <= maxT):
        curRes = [0] * len(init)
        curRes[0] = answer[-1][0]
        curRes[-1] = answer[-1][-1]
        for i in range(1, l - 1):
            A = answer[-1][i]
            B = -s * (answer[-1][i] - answer[-1][i - 1])
            C = r * (answer[-1][i - 1] - 2 * answer[-1][i] + answer[-1][i + 1])
            curRes[i] = A + B + C

        answer.append(curRes)
        curT += dt
    return answer
```

```python
def implicit_upstream(maxX, dx, maxT, dt, u, k, initAlgo):
    s = u * dt / dx
    r = k * dt / (dx * dx)
    
    init = initAlgo(maxX, dx)
    answer = []
    answer.append(init)
    curT = dt
    l = len(init)
    while (curT <= maxT):

        matrix = [0] * l
        for i in range(0, l):
            matrix[i] = [0] * 4

        matrix[0][1] = 1
        matrix[0][3] = answer[-1][0]

        matrix[-1][1] = 1
        matrix[-1][3] = answer[-1][-1]
        for i in range(1, l - 1):
            D = answer[-1][i] / dt
            A = -r
            B = 1 - s + 2 * r
            C = s - r
            matrix[i] = [A, B, C, D]
        answer.append(solve_system(matrix))
        curT += dt

    return answer
```

```python
def implicit_downstream(maxX, dx, maxT, dt, u, k, initAlgo):
    s = u * dt / dx
    r = k * dt / (dx * dx)
    
    init = initAlgo(maxX, dx)
    answer = []
    answer.append(init)
    curT = dt
    l = len(init)
    while (curT <= maxT):
        matrix = [0] * l
        for i in range(0, l):
            matrix[i] = [0] * 4

        matrix[0][1] = 1
        matrix[0][3] = answer[-1][0]

        matrix[-1][1] = 1
        matrix[-1][3] = answer[-1][-1]
        for i in range(1, l - 1):
            D = answer[-1][i]
            A = -s - r
            B = 1 + s + 2 * r
            C = -r
            matrix[i] = [A, B, C, D]
        answer.append(solve_system(matrix))
        curT += dt

    return answer
```

```python
def test_algo(algo, maxX=40, dx=1.0, maxT=100, dt=1.0, u=1.0, k=1.0, initAlgo=initHill, curt=0):
    graph = algo(k=k, dt=dt, dx=dx, maxT=maxT, u=u, maxX=maxX, initAlgo=initHill)
    xs = np.arange(-maxX, maxX+dx, dx)
    fig, ax = plt.subplots(figsize=(30,30))
    ax.plot(xs, graph[curt])
```

```python
test_algo(explicit_downstream,curt=53)
```

```python
test_algo(explicit_upstream,curt=70)
```

```python
test_algo(implicit_downstream,dt=0.5,curt=0)
```

```python
test_algo(implicit_upstream,curt=150,dt=0.5)
```

```python
test_algo(leapfrog,curt=0.5,maxT=200,k=0,dt=0.5,dx=1.0)
```
